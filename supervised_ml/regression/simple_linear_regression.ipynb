```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
```

```python
df = pd.read_csv(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# dependent variable/response/output/label/y
y = df['price'] 
```

```python
# independent variable/feture/input/x
x = df[['horsepower']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
trainx, testx, trainy, testy = train_test_split(x,y,test_size=.2,random_state=0)
```

```python
plt.scatter(trainx,trainy,c='red',alpha=0.5)
plt.show()
```

# Task 1
## write many programs to create the linear regression model between remaining numeric cols and price(y) and display the graphical output

```python
model = LinearRegression()
model.fit(trainx,trainy)
score = model.score(testx,testy) *100
print(f'the model achieved a score = {score}')
```

```python
ypred = model.predict(trainx)
```

```python
plt.scatter(trainx,trainy,c='red',alpha=0.5,label='original')
plt.plot(trainx,ypred,'g--',label='prediction')
plt.title('prediction')
plt.xlabel('horsepower')
plt.ylabel('price')
plt.legend()
plt.text(50,45000,f'R2={score}')
plt.show()
```
